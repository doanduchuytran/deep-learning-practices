```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class RegressionNN(nn.Module):

    def __init__(self):

        super().__init__()

        self.net = nn.Sequential(

            nn.Linear(8,64),
            nn.ReLU(),

            nn.Linear(64,32),
            nn.ReLU(),

            nn.Linear(32,1)
        )

    def forward(self,x):

        return self.net(x)
```

```python
model = RegressionNN()

criterion = nn.MSELoss()

optimizer = optim.Adam(model.parameters(),lr=0.001)
```
