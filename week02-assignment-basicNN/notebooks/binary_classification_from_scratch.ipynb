### Description

Neural Network Binary Classification (From Scratch)

Dataset: Breast Cancer Wisconsin Dataset

Goal:
Classify tumors as benign or malignant.

### Import Libraries

```python
import sys
import os

# Add the parent directory to Python path
sys.path.append(os.path.abspath(".."))
```

```python
import numpy as np
import matplotlib.pyplot as plt

from framework.layers import Linear
from framework.activations import ReLU, Sigmoid
from framework.losses import BinaryCrossEntropy
from framework.model import NeuralNetwork
from framework.optimizer import SGD
from framework.trainer import Trainer
from framework.metrics import binary_accuracy

from datasets.cancer_dataset import CancerDataset
```

### Load Dataset

```python
dataset = CancerDataset()

X_train, X_test, y_train, y_test = dataset.load()

print("Training samples:", X_train.shape)
print("Test samples:", X_test.shape)
```

### Build Neural Network

```python
model = NeuralNetwork()

model.add(Linear(30,32))
model.add(ReLU())

model.add(Linear(32,16))
model.add(ReLU())

model.add(Linear(16,1))
model.add(Sigmoid())
```

### Training Configuration

```python
loss = BinaryCrossEntropy()

optimizer = SGD(lr=0.01)

trainer = Trainer(model, loss, optimizer)
```

### Train Model

```python
losses = trainer.train(X_train,y_train,epochs=100)
```

### Plot Training Loss

```python
plt.plot(losses)

plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

### Evaluate Model

```python
y_pred = model.forward(X_test)

acc = binary_accuracy(y_pred, y_test)

print("Test Accuracy:", acc)
```
