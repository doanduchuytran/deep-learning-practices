### Description

Multiclass classification from scratch

Dataset: MNIST Dataset

Goal: classify digits from 0 to 9

### Import Libraries

```python
import sys
import os

# Add the parent directory to Python path
sys.path.append(os.path.abspath(".."))
```

```python
import numpy as np
import matplotlib.pyplot as plt

from framework.layers import Linear
from framework.activations import ReLU, Softmax
from framework.losses import CrossEntropy
from framework.model import NeuralNetwork
from framework.optimizer import SGD
from framework.trainer import Trainer
from framework.metrics import accuracy

from datasets.mnist_dataset import MNISTDataset
```

### Load Dataset

```python
dataset = MNISTDataset()

X_train, X_test, y_train, y_test = dataset.load()

print("Training samples:", X_train.shape)
print("Test samples:", X_test.shape)
```

### Build Neural Network

```python
model = NeuralNetwork()

model.add(Linear(784,128))
model.add(ReLU())

model.add(Linear(128,64))
model.add(ReLU())

model.add(Linear(64,10))
model.add(Softmax())
```

### Training Configuration

```python
loss = CrossEntropy()

optimizer = SGD(lr=0.01)

trainer = Trainer(model,loss,optimizer)
```

### Train Model

```python
losses = trainer.train(X_train,y_train,epochs=100)
```

### Plot Training Loss

```python
plt.plot(losses)

plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

### Evaluate Model

```python
y_pred = model.forward(X_test)

acc = accuracy(y_pred, y_test)

print("Test Accuracy:", acc)
```
